```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
totalplayers = len(purchase_data_df["SN"].unique())
totalplayers_display = pd.DataFrame({"Total Players": [totalplayers]})
totalplayers_display
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#basic calculations 
unique_items = len((purchase_data_df["Item ID"]).unique())

average_price = (purchase_data_df["Price"].mean())

number_of_purchases = (purchase_data_df["Purchase ID"].count())

total_revenue = (purchase_data_df["Price"].sum())

#Data frame containing above calculations
summary_df = pd.DataFrame({"Number of Unique Items":[unique_items],
                           "Average Price":[average_price], 
                           "Number of Purchases": [number_of_purchases], 
                           "Total Revenue": [total_revenue]})
#display dataframe
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Group data by gender
gender_group = purchase_data_df.groupby(["Gender"])
#filter out duplicates using nunique function
unique_gender_df = gender_group.nunique()

#Track total for each gender
total_gender = unique_gender_df["SN"].sum()

#Track count and calculate percentage of players
count = unique_gender_df["SN"].unique()
percentage = unique_gender_df["SN"]/ total_gender

#
gender_df = pd.DataFrame({"Percentage of Players": percentage,
                            "Total Count":count})

#format as a percentage out to 2 decimal points using .map function
gender_df["Percentage of Players"] = gender_df["Percentage of Players"].map("{:,.2%}".format)

#display dataframe
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Purchase Count,Average Purchase Price,Total Purchase Value,Average Purchase Total per Person by Gender
#Count, mean, and sum functions used to find first three values
purchase_count = gender_group["Gender"].count()
average_purchase_price = gender_group["Price"].mean()
total_purchase_value = gender_group["Price"].sum()
#divide total purchase value by number of members in gender group
average_total_purchase_per_person = total_purchase_value / total_gender
#create dataframe with above values
purchasing_analysis_gender_df = pd.DataFrame({"Purchase Count": purchase_count,
                                              "Average Purchase Price": average_purchase_price,
                                              "Total Purchase Value": total_purchase_value,
                                              "Avg Total Purchase per Person": average_total_purchase_per_person                                             
                                              })
#sets index to "Gender" displaying above first column
purchasing_analysis_gender_df.index.name = "Gender"

#display dataframe
purchasing_analysis_gender_df

#Formatting last 3 columns out to 2 decimal places with $ signs
purchasing_analysis_gender_df.style.format({"Average Purchase Price":"${:,.2f}",
                                  "Total Purchase Value":"${:,.2f}",
                                  "Avg Total Purchase per Person":"${:,.2f}"})

```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python


#Establish bins with set age intervals 
bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 200]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Add new column with the age group each player falls into
purchase_data_df["Age Group"] = pd.cut(purchase_data_df["Age"], bins, labels=labels)


#New data frame with age group column 
age_group_df = purchase_data_df.groupby("Age Group")

#Find the total count for each age group to display in data frame and calculate percentage
age_total_count = age_group_df["SN"].nunique()

#Find percentage of each bin by dividing by total player count 
age_percentage_of_players = (age_total_count/totalplayers) * 100

#Display new data frame with total count per age bins and percentage of total players
age_demographics_df = pd.DataFrame({"Total Count":age_total_count,
                                    "Percentage of Players":age_percentage_of_players
                                    })
age_demographics_df

#Format percentage column to two decimal places and add % symbol
age_demographics_df.style.format({"Percentage of Players":"{:,.2f}%"})
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Purchase Count,Average Purchase Price,Total Purchase Value,Average Purchase Total per Person by Age Group
#Use above data frame grouped by age bins for calculations
#counting each purchase within age bins
purchase_count_age = age_group_df["Purchase ID"].count()
#average within price column for each bin
average_purchase_price_age = age_group_df["Price"].mean()
#sum the values within price for each bin
total_purchase_value_age = age_group_df["Price"].sum()

#caluclate average using total purchase value divided with the count for each age bin
average_total_purchase_per_person = total_purchase_value_age/age_total_count

#purchasing anylsis by age data frame displaying above values
purchasing_analysis_age_df=pd.DataFrame({"Purchase Count":purchase_count_age,
                                         "Average Purchase Price":average_purchase_price_age,
                                         "Total Purchase Value":total_purchase_value_age, 
                                         "Avg Total Purchase per Person":average_total_purchase_per_person
                                        })
purchasing_analysis_age_df

#format latter three columns out to 2 decimal places and include $ sign
purchasing_analysis_age_df.style.format({"Average Purchase Price":"${:,.2f}",
                               "Total Purchase Value":"${:,.2f}",
                               "Avg Total Purchase per Person":"${:,.2f}"})
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#SN,Purchase Count,Average Purchase Price,Total Purchase Value
#grouping by individual screen names
sn_stats = purchase_data_df.groupby("SN")
#counting purchases for each spender
purchase_count_spenders = sn_stats["Purchase ID"].count()
#grabbing average from price column
average_purchase_price_spenders = sn_stats["Price"].mean()
#summing price column 
total_purchase_value_spenders = sn_stats["Price"].sum()

#dataframe displaying above values
top_spenders_df = pd.DataFrame({"Purchase Count":purchase_count_spenders,
                                "Average Purchase Price":average_purchase_price_spenders,
                                "Total Purchase Value":total_purchase_value_spenders
                                })
#Display spenders in descending orde of total purchase value to identify top spenders
top_spenders_ordered_df = top_spenders_df.sort_values(["Total Purchase Value"], ascending=False)
top_spenders_ordered_df

#Format monetary values out to two decimal places with $ symbol
top_spenders_ordered_df.style.format({"Average Purchase Price":"${:,.2f}",
                                      "Total Purchase Value":"${:,.2f}"
                                     })
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Item ID,Item Name,Purchase Count,Item Price,Total Purchase Value
#retrieving item id, item name, and price columns
item_stats_df = purchase_data_df.groupby(["Item ID","Item Name","Price"])

#counts number of times item was purchased
purchase_count_item = item_stats_df["Price"].count()
#Total spent on each item using sum function 
total_purchase_value_item = item_stats_df["Price"].sum()
#find price of each item by dividing total spent by instances it was purchased
item_price = total_purchase_value_item/purchase_count_item

#Create dataframe to display values 
most_popular_item_df = pd.DataFrame({"Purchase Count":purchase_count_item,
                                     "Item Price":item_price,
                                     "Total Purchase Value":total_purchase_value_item
                                    })

#Order dataframe by purchase count to isolate most purchased items
most_popular_item_ordered_df = most_popular_item_df.sort_values(["Purchase Count"],ascending=False)
most_popular_item_ordered_df
#format monetary values out to two decimal places and include $ sign
most_popular_item_ordered_df.style.format({"Item Price":"${:,.2f}",
                                "Total Purchase Value":"${:,.2f}"
                                          })
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Sort Total Purchase Value column in descending order
most_profitable_items_df = most_popular_item_ordered_df.sort_values(["Total Purchase Value"],ascending=False)
most_profitable_items_df
#Format monetary values
most_profitable_items_df.style.format({"Item Price":"${:,.2f}",
                                "Total Purchase Value":"${:,.2f}"
                                      })
```
